```python
# ==============================================================================
# SNIPPET 1: CONFIGURACIÓN Y PREPARACIÓN CON PLANIFICADORES
# ==============================================================================

from datetime import datetime
from pyspark.sql.functions import (
    col, year, month, max as spark_max, lit, substring, coalesce, 
    sum as spark_sum, lag, when, round as spark_round, countDistinct, count
)
import pyspark.sql.functions as F
from pyspark.sql.window import Window

print("=== SNIPPET 1: CONFIGURACIÓN Y PREPARACIÓN ===")

# CONFIGURACIÓN GLOBAL
fecha_actual = datetime.now()
ano_actual = fecha_actual.year
mes_actual_num = fecha_actual.month

nombres_meses = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril", 5: "Mayo", 6: "Junio",
    7: "Julio", 8: "Agosto", 9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

meses_mapping = {
    1: ("ENE", "Enero"), 2: ("FEB", "Febrero"), 3: ("MAR", "Marzo"), 
    4: ("ABR", "Abril"), 5: ("MAY", "Mayo"), 6: ("JUN", "Junio"),
    7: ("JUL", "Julio"), 8: ("AGO", "Agosto"), 9: ("SEP", "Septiembre"), 
    10: ("OCT", "Octubre"), 11: ("NOV", "Noviembre"), 12: ("DIC", "Diciembre")
}

mes_ejecucion_actual = nombres_meses[mes_actual_num]
anos_a_procesar = list(range(ano_actual, ano_actual + 6))

# Extraer SKUs válidos
skus_validos = []
for region, paises in dict_skus.items():
    for pais, familias in paises.items():
        for familia, skus in familias.items():
            skus_validos.extend(skus)

paises_requeridos = ["CL", "CO", "EC", "PE", "CR", "GT", "HN", "SV","AR","CA","MX"]

print(f"Año: {ano_actual} | Mes: {mes_ejecucion_actual} | SKUs: {len(skus_validos):,}")

# CARGA DE DATOS
print("\n--- Cargando tablas ---")
df_demanda = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_demanda")
df_entregas = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_entregas")
df_costeado = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_costeado")
df_compras = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_compras")
df_facturacion = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_facturacion")

# MESES Y AÑOS
meses_data = [(mes_num, codigo, nombre, mes_num) for mes_num, (codigo, nombre) in meses_mapping.items()]
df_meses = spark.createDataFrame(meses_data, ["ORDENM", "Attribute", "Mes_nombre", "mes_numero"])
df_anos = spark.createDataFrame([(ano,) for ano in anos_a_procesar], ["Ano"])

# ===============================================================================
# DEMANDA CON PLANIFICADOR + COMPLEMENTO CSV 2026
# ===============================================================================

print("\n" + "="*80)
print("=== PREPARANDO DEMANDA CON PLANIFICADOR + COMPLEMENTO CSV 2026 ===")
print("="*80)

# -------------------------------------------------------------------------------
# PASO 1: CARGAR DEMANDA DE CUBOS (FUENTE PRINCIPAL) - POR PLANIFICADOR
# -------------------------------------------------------------------------------

print("\n→ PASO 1: Cargando demanda desde CUBOS (fuente principal)...")

df_demanda_prep = df_demanda.filter(
    (substring(col("Ejercicio_Periodo"), 1, 4).cast("int") >= ano_actual) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Pais").isNotNull()) &
    (col("Material").isin(skus_validos)) &
    (col("Grupo_Articulos") != "-") &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Planificador_Necesidades").alias("Planificador"),
    col("Material"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
)

df_demanda_cubos = df_demanda_prep.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(
    F.sum("Fcst").alias("Fcst_cubos"),
    F.count("Material").alias("num_materiales_cubos")
).withColumn(
    "Planificador",
    when(col("Planificador").isNull(), lit("SIN_PLANIFICADOR")).otherwise(col("Planificador"))
)

print(f"  ✓ Demanda desde CUBOS: {df_demanda_cubos.count():,} registros")

# -------------------------------------------------------------------------------
# PASO 2: CARGAR CSV TEMPORAL 2026 (COMPLEMENTO) - POR FAMILIA (SIN PLANIFICADOR)
# -------------------------------------------------------------------------------

csv_cargado = False

print("\n→ PASO 2: Intentando cargar CSV temporal 2026 (complemento)...")

try:
    df_csv_2026_raw = spark.read.format("csv") \
        .option("header", "true") \
        .option("inferSchema", "true") \
        .load("Files/BGT_2026.csv")
    
    print("  ✓ CSV cargado desde Files/BGT_2026.csv")
    csv_cargado = True
    
    print(f"  → Columnas del CSV: {df_csv_2026_raw.columns}")
    print(f"  → Registros crudos: {df_csv_2026_raw.count():,}")
    
    df_csv_prep = df_csv_2026_raw.filter(
        (col("País").isin(paises_requeridos)) &
        (col("Ejercicio_Periodo").isNotNull()) &
        (col("Grupo_Articulos").isNotNull())
    ).select(
        col("País").alias("Pais"),
        col("Grupo_Articulos").alias("Familia"),
        col("Ejercicio_Periodo"),
        col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
    )
    
    df_csv_prep = df_csv_prep.withColumn(
        "Ano", (col("Ejercicio_Periodo") / 1000).cast("int")
    ).withColumn(
        "mes_numero", (col("Ejercicio_Periodo") % 1000).cast("int")
    ).filter(
        (col("Ano") == 2026) & (col("mes_numero").between(1, 12))
    )
    
    df_demanda_csv = df_csv_prep.groupBy(
        "Pais", "Familia", "Ano", "mes_numero"
    ).agg(
        F.sum("Fcst").alias("Fcst_csv"),
        F.count("*").alias("num_registros_csv")
    )
    
    print(f"  ✓ Demanda desde CSV 2026 agregada: {df_demanda_csv.count():,} registros")
    
except Exception as e:
    print(f"\n  ⚠️  Error al cargar CSV: {str(e)}")
    csv_cargado = False

# -------------------------------------------------------------------------------
# PASO 3: COMBINAR CUBOS + CSV (PRIORIDAD A CUBOS CON VALOR REAL > 0)
# -------------------------------------------------------------------------------

if csv_cargado:
    print("\n→ PASO 3: Combinando CUBOS + CSV (prioridad a CUBOS con valores REALES)...")
    print("  🔑 LÓGICA: Si cubos > 0 → CUBOS, si cubos = 0 o NULL → CSV de la familia")
    
    df_demanda_completa = df_demanda_cubos.alias("cubos").join(
        df_demanda_csv.alias("csv"),
        (col("cubos.Pais") == col("csv.Pais")) &
        (col("cubos.Familia") == col("csv.Familia")) &
        (col("cubos.Ano") == col("csv.Ano")) &
        (col("cubos.mes_numero") == col("csv.mes_numero")),
        "left"
    ).select(
        col("cubos.Pais").alias("Pais"),
        col("cubos.Familia").alias("Familia"),
        col("cubos.Planificador").alias("Planificador"),
        col("cubos.Ano").alias("Ano"),
        col("cubos.mes_numero").alias("mes_numero"),
        when(
            (col("cubos.Fcst_cubos").isNotNull()) & (col("cubos.Fcst_cubos") > 0),
            col("cubos.Fcst_cubos")
        ).otherwise(
            coalesce(col("csv.Fcst_csv"), lit(0.0))
        ).alias("Fcst_total")
    )
    
    print(f"  ✓ Demanda TOTAL combinada: {df_demanda_completa.count():,} registros")
    
    df_demanda_agg = df_demanda_completa.select(
        "Pais", "Familia", "Planificador", "Ano", "mes_numero", "Fcst_total"
    )
    
    print("  ✅ Demanda combinada CUBOS+CSV lista")
    
else:
    print("\n→ PASO 3: Usando solo datos de CUBOS (CSV no disponible)...")
    df_demanda_agg = df_demanda_cubos.withColumnRenamed("Fcst_cubos", "Fcst_total")
    print("  ✅ Demanda desde CUBOS únicamente")

# MATERIALIZAR DEMANDA
df_demanda_agg.cache()
df_demanda_agg.count()
print(f"✓ Demanda materializada: {df_demanda_agg.count():,} registros")

# FACTURACIÓN CON PLANIFICADOR
print("\n--- Preparando Facturación con Planificador ---")
df_facturacion_prep = df_facturacion.filter(
    (col("Material").isin(skus_validos)) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Planificador_Necesidades").alias("Planificador"),
    col("Material"),
    col("Volumen_ventas").alias("Facturacion"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero")
).filter(col("Ano") >= ano_actual)

df_facturacion_agg = df_facturacion_prep.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(
    F.sum("Facturacion").alias("Facturacion_total")
).withColumn(
    "Planificador",
    when(col("Planificador").isNull(), lit("SIN_PLANIFICADOR")).otherwise(col("Planificador"))
).cache()

print(f"Facturación por planificador: {df_facturacion_agg.count():,}")
print("[COMPLETADO] Snippet 1")
```

```python
# ==============================================================================
# SNIPPET 2: LLEGADAS E INVENTARIOS CON PLANIFICADORES
# ==============================================================================

print("\n=== SNIPPET 2: LLEGADAS E INVENTARIOS ===")

# ENTREGAS CON PLANIFICADOR
print("\n--- Preparando Entregas ---")
df_entregas_prep = df_entregas.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_entrega_confirmada")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Planificador_de_Necesidades").alias("Planificador"),
    col("Material"),
    year(col("Fecha_entrega_confirmada")).alias("Ano"),
    month(col("Fecha_entrega_confirmada")).alias("mes_numero"),
    col("Cantidad_Entregada_Efectivamente").alias("Entregas")
)

df_entregas_agg = df_entregas_prep.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(F.sum("Entregas").alias("Entregas_total"))

# COMPRAS CON PLANIFICADOR
print("--- Preparando Compras ---")
df_compras_prep = df_compras.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_Final")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Planificador_Necesidades").alias("Planificador"),
    col("Material"),
    year(col("Fecha_Final")).alias("Ano"),
    month(col("Fecha_Final")).alias("mes_numero"),
    col("Qty_sin_EE").alias("Compras")
)

df_compras_agg = df_compras_prep.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(F.sum("Compras").alias("Compras_total"))

# COMBINAR LLEGADAS
df_llegadas_combinado = df_entregas_agg.select(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero",
    col("Entregas_total").alias("Cantidad")
).union(
    df_compras_agg.select(
        "Pais", "Familia", "Planificador", "Ano", "mes_numero",
        col("Compras_total").alias("Cantidad")
    )
)

df_llegadas_agg = df_llegadas_combinado.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(
    F.sum("Cantidad").alias("Llegadas_total")
).withColumn(
    "Planificador",
    when(col("Planificador").isNull(), lit("SIN_PLANIFICADOR")).otherwise(col("Planificador"))
).cache()

print(f"Llegadas por planificador: {df_llegadas_agg.count():,}")

# INVENTARIOS - Verificar si tiene planificador
print("\n--- Procesando Inventarios ---")
planif_col_costeado = None
for col_name in df_costeado.columns:
    if "planificador" in col_name.lower():
        planif_col_costeado = col_name
        break

if planif_col_costeado:
    print(f"Costeado con planificador: {planif_col_costeado}")
    df_inventarios_por_semana = df_costeado.filter(
        (col("Sociedad_Pais").isin(paises_requeridos)) &
        (col("Grupo_Articulos").isNotNull()) &
        (col("Grupo_Articulos") != "-") &
        (substring(col("Ejercicio___Periodo"), 1, 4).cast("int") >= ano_actual - 1)
    ).groupBy(
        col("Sociedad_Pais").alias("Pais"),
        col("Grupo_Articulos").alias("Familia"),
        col(planif_col_costeado).alias("Planificador"),
        substring(col("Ejercicio___Periodo"), 1, 4).cast("int").alias("Ano"),
        substring(col("Ejercicio___Periodo"), 5, 3).cast("int").alias("mes_numero"),
        col("Semana")
    ).agg(F.sum("Cantidad_Stock_Total_").alias("Inventario_semana"))
else:
    print("Costeado SIN planificador - cross join")
    planificadores_unicos = df_demanda_agg.select("Planificador").distinct()
    
    df_inv_base = df_costeado.filter(
        (col("Sociedad_Pais").isin(paises_requeridos)) &
        (col("Grupo_Articulos").isNotNull()) &
        (col("Grupo_Articulos") != "-") &
        (substring(col("Ejercicio___Periodo"), 1, 4).cast("int") >= ano_actual - 1)
    ).groupBy(
        col("Sociedad_Pais").alias("Pais"),
        col("Grupo_Articulos").alias("Familia"),
        substring(col("Ejercicio___Periodo"), 1, 4).cast("int").alias("Ano"),
        substring(col("Ejercicio___Periodo"), 5, 3).cast("int").alias("mes_numero"),
        col("Semana")
    ).agg(F.sum("Cantidad_Stock_Total_").alias("Inventario_semana"))
    
    df_inventarios_por_semana = df_inv_base.crossJoin(planificadores_unicos)

df_inventarios_por_semana.cache()
print(f"Inventarios por semana: {df_inventarios_por_semana.count():,}")

# INVENTARIO FINAL E INICIAL
df_limites_semanas = df_inventarios_por_semana.groupBy(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero"
).agg(spark_max("Semana").alias("ultima_semana"))

df_inv_final = df_inventarios_por_semana.alias("inv").join(
    df_limites_semanas.alias("lim"),
    (col("inv.Pais") == col("lim.Pais")) &
    (col("inv.Familia") == col("lim.Familia")) &
    (col("inv.Planificador") == col("lim.Planificador")) &
    (col("inv.Ano") == col("lim.Ano")) &
    (col("inv.mes_numero") == col("lim.mes_numero")) &
    (col("inv.Semana") == col("lim.ultima_semana")),
    "inner"
).select(
    col("inv.Pais"), col("inv.Familia"), col("inv.Planificador"),
    col("inv.Ano"), col("inv.mes_numero"),
    col("inv.Inventario_semana").alias("Inv_Final_Mes")
)

windowSpec = Window.partitionBy("Pais", "Familia", "Planificador").orderBy("Ano", "mes_numero")

df_inventarios_con_inicial = df_inv_final.withColumn(
    "Inv_Inicial_Mes", 
    lag(col("Inv_Final_Mes"), 1).over(windowSpec)
).withColumn(
    "Inv_Inicial", 
    when(col("Inv_Inicial_Mes").isNull(), lit(0.0)).otherwise(col("Inv_Inicial_Mes"))
).select(
    "Pais", "Familia", "Planificador", "Ano", "mes_numero",
    "Inv_Inicial",
    col("Inv_Final_Mes").alias("Inv_Final")
).filter(col("Ano") >= ano_actual).cache()

print(f"Inventarios por planificador: {df_inventarios_con_inicial.count():,}")
print("[COMPLETADO] Snippet 2")
```

```python
# ==============================================================================
# SNIPPET 3: ESTRUCTURA FINAL SOLO PLANIFICADORES
# ==============================================================================

print("\n=== SNIPPET 3: ESTRUCTURA FINAL ===")

# ESTRUCTURA BASE
print("\n--- Creando estructura base ---")
paises_familias_planif = df_inventarios_con_inicial.select("Pais", "Familia", "Planificador").distinct()

df_inventarios_master = paises_familias_planif \
    .crossJoin(df_anos) \
    .crossJoin(df_meses) \
    .alias("base") \
    .join(
        df_inventarios_con_inicial.alias("inv"),
        (col("base.Pais") == col("inv.Pais")) &
        (col("base.Familia") == col("inv.Familia")) &
        (col("base.Planificador") == col("inv.Planificador")) &
        (col("base.Ano") == col("inv.Ano")) &
        (col("base.ORDENM") == col("inv.mes_numero")),
        "left"
    ).select(
        col("base.Ano"),
        lit(mes_ejecucion_actual).alias("Mes"),
        col("base.Pais").alias("PAIS"),
        col("base.Familia"),
        col("base.Planificador"),
        col("base.Attribute"),
        col("base.ORDENM"),
        lit(1).alias("PWOH"),
        coalesce(col("inv.Inv_Inicial"), lit(0.0)).alias("Inv_Inicial"),
        coalesce(col("inv.Inv_Final"), lit(0.0)).alias("Inv_Final")
    ).cache()

print(f"Estructura base: {df_inventarios_master.count():,}")

# JOINS
print("\n--- Uniendo métricas ---")

df_con_demanda = df_inventarios_master.alias("base").join(
    df_demanda_agg.alias("dem"),
    (col("base.PAIS") == col("dem.Pais")) &
    (col("base.Familia") == col("dem.Familia")) &
    (col("base.Planificador") == col("dem.Planificador")) &
    (col("base.Ano") == col("dem.Ano")) &
    (col("base.ORDENM") == col("dem.mes_numero")),
    "left"
).select(col("base.*"), coalesce(col("dem.Fcst_total"), lit(0.0)).alias("Fcst"))

df_con_facturacion = df_con_demanda.alias("base").join(
    df_facturacion_agg.alias("fact"),
    (col("base.PAIS") == col("fact.Pais")) &
    (col("base.Familia") == col("fact.Familia")) &
    (col("base.Planificador") == col("fact.Planificador")) &
    (col("base.Ano") == col("fact.Ano")) &
    (col("base.ORDENM") == col("fact.mes_numero")),
    "left"
).select(col("base.*"), coalesce(col("fact.Facturacion_total"), lit(0.0)).alias("Facturacion"))

df_master_planif = df_con_facturacion.alias("base").join(
    df_llegadas_agg.alias("lleg"),
    (col("base.PAIS") == col("lleg.Pais")) &
    (col("base.Familia") == col("lleg.Familia")) &
    (col("base.Planificador") == col("lleg.Planificador")) &
    (col("base.Ano") == col("lleg.Ano")) &
    (col("base.ORDENM") == col("lleg.mes_numero")),
    "left"
).select(
    col("base.Ano"), col("base.Mes"), col("base.PAIS"), col("base.Familia"),
    col("base.Planificador"), col("base.Attribute"), col("base.ORDENM"), col("base.PWOH"),
    col("base.Inv_Inicial"), col("base.Inv_Final"),
    coalesce(col("lleg.Llegadas_total"), lit(0.0)).alias("Llegadas"),
    col("base.Fcst"), col("base.Facturacion")
)

print(f"Tabla por planificador: {df_master_planif.count():,}")

# GUARDAR
print("\n--- Guardando tabla ---")
spark.catalog.clearCache()

df_master_optimizado = df_master_planif.repartition(150, "PAIS", "Familia", "Planificador")

spark.sql("USE LH_CARATULAS_PLNF")
spark.sql("DROP TABLE IF EXISTS continental_master_planificadores")

df_master_optimizado.write.format("delta").mode("overwrite").saveAsTable("continental_master_planificadores")

print("\n✅ TABLA GUARDADA: continental_master_planificadores")

# VERIFICACIÓN
print("\n--- Verificación CR-LAVA por planificador ---")
spark.sql("""
    SELECT Planificador, Attribute, ORDENM,
           ROUND(Inv_Inicial, 1) as Inv_Inicial,
           ROUND(Inv_Final, 1) as Inv_Final,
           ROUND(Llegadas, 1) as Llegadas,
           ROUND(Fcst, 1) as Fcst
    FROM LH_CARATULAS_PLNF.continental_master_planificadores
    WHERE PAIS = 'CR' AND Familia = 'LAVA' AND Ano = 2026
    ORDER BY Planificador, ORDENM
    LIMIT 20
""").show(20, truncate=False)

print("\n✅ ETL PLANIFICADORES COMPLETADO")
print("[COMPLETADO] Snippet 3")
```

```python
# ==============================================================================
# SNIPPET 3: ESTRUCTURA FINAL SOLO PLANIFICADORES (CORREGIDO)
# ==============================================================================
# FIX: El join de demanda ahora hace dos LEFT JOINs separados:
#   1) df_demanda_cubos (con Planificador) - datos reales de cubos
#   2) df_demanda_csv (sin Planificador) - complemento CSV 2026
# Esto asegura que los meses donde un planificador NO tiene filas en cubos
# reciban el S&OP del CSV como fallback, en vez de quedar en 0.
# ==============================================================================

print("\n=== SNIPPET 3: ESTRUCTURA FINAL ===")

# ESTRUCTURA BASE
print("\n--- Creando estructura base ---")
paises_familias_planif = df_inventarios_con_inicial.select("Pais", "Familia", "Planificador").distinct()

df_inventarios_master = paises_familias_planif \
    .crossJoin(df_anos) \
    .crossJoin(df_meses) \
    .alias("base") \
    .join(
        df_inventarios_con_inicial.alias("inv"),
        (col("base.Pais") == col("inv.Pais")) &
        (col("base.Familia") == col("inv.Familia")) &
        (col("base.Planificador") == col("inv.Planificador")) &
        (col("base.Ano") == col("inv.Ano")) &
        (col("base.ORDENM") == col("inv.mes_numero")),
        "left"
    ).select(
        col("base.Ano"),
        lit(mes_ejecucion_actual).alias("Mes"),
        col("base.Pais").alias("PAIS"),
        col("base.Familia"),
        col("base.Planificador"),
        col("base.Attribute"),
        col("base.ORDENM"),
        lit(1).alias("PWOH"),
        coalesce(col("inv.Inv_Inicial"), lit(0.0)).alias("Inv_Inicial"),
        coalesce(col("inv.Inv_Final"), lit(0.0)).alias("Inv_Final")
    ).cache()

print(f"Estructura base: {df_inventarios_master.count():,}")

# JOINS
print("\n--- Uniendo métricas ---")

# =========================================================================
# DEMANDA - FIX: Dos LEFT JOINs separados (cubos + CSV)
# =========================================================================
# Lógica: Si cubos tiene forecast > 0 para ese planificador → usa cubos
#          Si no → usa el CSV de la familia (que no tiene planificador)
# Esto garantiza que TODOS los meses de TODOS los planificadores tengan S&OP

if csv_cargado:
    print("  → Demanda: Combinando CUBOS (por planificador) + CSV (por familia)...")
    df_con_demanda = df_inventarios_master.alias("base").join(
        df_demanda_cubos.alias("cubos"),
        (col("base.PAIS") == col("cubos.Pais")) &
        (col("base.Familia") == col("cubos.Familia")) &
        (col("base.Planificador") == col("cubos.Planificador")) &
        (col("base.Ano") == col("cubos.Ano")) &
        (col("base.ORDENM") == col("cubos.mes_numero")),
        "left"
    ).join(
        df_demanda_csv.alias("csv"),
        (col("base.PAIS") == col("csv.Pais")) &
        (col("base.Familia") == col("csv.Familia")) &
        (col("base.Ano") == col("csv.Ano")) &
        (col("base.ORDENM") == col("csv.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        when(
            (col("cubos.Fcst_cubos").isNotNull()) & (col("cubos.Fcst_cubos") > 0),
            col("cubos.Fcst_cubos")
        ).otherwise(
            coalesce(col("csv.Fcst_csv"), lit(0.0))
        ).alias("Fcst")
    )
else:
    print("  → Demanda: Solo CUBOS (CSV no disponible)...")
    df_con_demanda = df_inventarios_master.alias("base").join(
        df_demanda_cubos.alias("cubos"),
        (col("base.PAIS") == col("cubos.Pais")) &
        (col("base.Familia") == col("cubos.Familia")) &
        (col("base.Planificador") == col("cubos.Planificador")) &
        (col("base.Ano") == col("cubos.Ano")) &
        (col("base.ORDENM") == col("cubos.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("cubos.Fcst_cubos"), lit(0.0)).alias("Fcst")
    )

# =========================================================================
# FACTURACIÓN (sin cambios - ya funciona bien)
# =========================================================================
df_con_facturacion = df_con_demanda.alias("base").join(
    df_facturacion_agg.alias("fact"),
    (col("base.PAIS") == col("fact.Pais")) &
    (col("base.Familia") == col("fact.Familia")) &
    (col("base.Planificador") == col("fact.Planificador")) &
    (col("base.Ano") == col("fact.Ano")) &
    (col("base.ORDENM") == col("fact.mes_numero")),
    "left"
).select(col("base.*"), coalesce(col("fact.Facturacion_total"), lit(0.0)).alias("Facturacion"))

# =========================================================================
# LLEGADAS (sin cambios - ya funciona bien)
# =========================================================================
df_master_planif = df_con_facturacion.alias("base").join(
    df_llegadas_agg.alias("lleg"),
    (col("base.PAIS") == col("lleg.Pais")) &
    (col("base.Familia") == col("lleg.Familia")) &
    (col("base.Planificador") == col("lleg.Planificador")) &
    (col("base.Ano") == col("lleg.Ano")) &
    (col("base.ORDENM") == col("lleg.mes_numero")),
    "left"
).select(
    col("base.Ano"), col("base.Mes"), col("base.PAIS"), col("base.Familia"),
    col("base.Planificador"), col("base.Attribute"), col("base.ORDENM"), col("base.PWOH"),
    col("base.Inv_Inicial"), col("base.Inv_Final"),
    coalesce(col("lleg.Llegadas_total"), lit(0.0)).alias("Llegadas"),
    col("base.Fcst"), col("base.Facturacion")
)

print(f"Tabla por planificador: {df_master_planif.count():,}")

# =========================================================================
# VERIFICACIÓN DEL FIX - Comprobar que S&OP llega a todos los meses
# =========================================================================
print("\n--- Verificación de cobertura S&OP por planificador ---")
df_master_planif.filter(
    (col("PAIS") == "CR") & (col("Familia") == "LAVA") & (col("Ano") == 2026)
).groupBy("Planificador").agg(
    F.count(when(col("Fcst") > 0, True)).alias("Meses_con_SOP"),
    F.count("*").alias("Meses_totales"),
    F.sum("Fcst").alias("Fcst_total")
).orderBy("Planificador").show(20, truncate=False)

# GUARDAR
print("\n--- Guardando tabla ---")
spark.catalog.clearCache()

df_master_optimizado = df_master_planif.repartition(150, "PAIS", "Familia", "Planificador")

spark.sql("USE LH_CARATULAS_PLNF")
spark.sql("DROP TABLE IF EXISTS continental_master_planificadores")

df_master_optimizado.write.format("delta").mode("overwrite").saveAsTable("continental_master_planificadores")

print("\n✅ TABLA GUARDADA: continental_master_planificadores")

# VERIFICACIÓN
print("\n--- Verificación CR-LAVA por planificador ---")
spark.sql("""
    SELECT Planificador, Attribute, ORDENM,
           ROUND(Inv_Inicial, 1) as Inv_Inicial,
           ROUND(Inv_Final, 1) as Inv_Final,
           ROUND(Llegadas, 1) as Llegadas,
           ROUND(Fcst, 1) as Fcst
    FROM LH_CARATULAS_PLNF.continental_master_planificadores
    WHERE PAIS = 'CR' AND Familia = 'LAVA' AND Ano = 2026
    ORDER BY Planificador, ORDENM
    LIMIT 24
""").show(24, truncate=False)

print("\n✅ ETL PLANIFICADORES COMPLETADO")
print("[COMPLETADO] Snippet 3")
```
