```python
#!/usr/bin/env python
# coding: utf-8

# ## NB_ETL_DAILY_CARATULAS - VERSIÓN CORREGIDA
# 
# Corrección fundamental en inventarios:
# - Inv_Inicial: Última semana del mes ANTERIOR
# - Inv_Final: Última semana del mes ACTUAL

# ===============================================================================
# NOTEBOOK 1: CONTINENTAL_MASTER_ETL_DAILY - CORREGIDO
# Actualización diaria - Solo procesamiento básico y estable
# ===============================================================================

from datetime import datetime
from pyspark.sql.functions import col, year, month, max as spark_max, min as spark_min, count, lit, substring, coalesce, sum as spark_sum, lag, when, isnan, isnull
from pyspark.sql.types import *
import pyspark.sql.functions as F
from pyspark.sql.window import Window

print("=== CONTINENTAL MASTER ETL DAILY - CORREGIDO ===")

# ===============================================================================
# 1. CONFIGURACIÓN GLOBAL (EXACTA DE TU CÓDIGO)
# ===============================================================================

# dict_skus ya debe estar disponible desde la celda anterior

# PARAMETROS GLOBALES (exacto de tu código)
fecha_actual = datetime.now()
ano_actual = fecha_actual.year
mes_actual_num = fecha_actual.month

# Mapeo de meses
nombres_meses = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril", 5: "Mayo", 6: "Junio",
    7: "Julio", 8: "Agosto", 9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

meses_mapping = {
    1: ("ENE", "Enero"), 2: ("FEB", "Febrero"), 3: ("MAR", "Marzo"), 
    4: ("ABR", "Abril"), 5: ("MAY", "Mayo"), 6: ("JUN", "Junio"),
    7: ("JUL", "Julio"), 8: ("AGO", "Agosto"), 9: ("SEP", "Septiembre"), 
    10: ("OCT", "Octubre"), 11: ("NOV", "Noviembre"), 12: ("DIC", "Diciembre")
}

mes_ejecucion_actual = nombres_meses[mes_actual_num]
anos_a_procesar = list(range(ano_actual, ano_actual + 6))

# Extraer SKUs válidos
skus_validos = []
for region, paises in dict_skus.items():
    for pais, familias in paises.items():
        for familia, skus in familias.items():
            skus_validos.extend(skus)

paises_requeridos = ["CR","GT","HN","SV","PE","CO","EC","CL","AR","CA"]

print(f"Fecha actual del sistema: {fecha_actual.strftime('%Y-%m-%d')}")
print(f"Año actual: {ano_actual}")
print(f"Mes de ejecución: {mes_actual_num} ({mes_ejecucion_actual})")
print(f"Años a procesar: {anos_a_procesar}")
print(f"Total SKUs válidos: {len(skus_validos)}")
print(f"Países requeridos: {paises_requeridos}")

# ===============================================================================
# 2. CARGA DE DATOS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== CARGANDO TABLAS DEL LAKEHOUSE ===")

df_demanda = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_demanda")
df_entregas = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_entregas") 
df_costeado = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_costeado")
df_calendario = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_calendario")
df_compras = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_compras")
df_facturacion = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_facturacion")

print("Tablas cargadas exitosamente")
print(f"  - continental_demanda: {df_demanda.count():,} registros")
print(f"  - continental_entregas: {df_entregas.count():,} registros")
print(f"  - continental_costeado: {df_costeado.count():,} registros")
print(f"  - continental_compras: {df_compras.count():,} registros")
print(f"  - continental_facturacion: {df_facturacion.count():,} registros")

# ===============================================================================
# 3. PREPARACIÓN DE DEMANDA (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO DATOS DE DEMANDA ===")

# Crear DataFrame de meses para cross join
meses_data = [(mes_num, codigo, nombre, mes_num) for mes_num, (codigo, nombre) in meses_mapping.items()]
df_meses = spark.createDataFrame(meses_data, ["ORDENM", "Attribute", "Mes_nombre", "mes_numero"])
df_anos = spark.createDataFrame([(ano,) for ano in anos_a_procesar], ["Ano"])

# Filtrar demanda solo con SKUs válidos desde año actual
df_demanda_prep = df_demanda.filter(
    (substring(col("Ejercicio_Periodo"), 1, 4).cast("int") >= ano_actual) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Pais").isNotNull()) &
    (col("Material").isin(skus_validos)) &
    (col("Grupo_Articulos") != "-")
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
)

# Agrupar demanda por dimensiones principales
df_demanda_agg = df_demanda_prep.filter(
    col("Pais").isin(paises_requeridos)
).groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Fcst").alias("Fcst_total"),
    F.count("Material").alias("num_materiales")
)

print(f"Demanda agregada: {df_demanda_agg.count():,} registros")

# MATERIALIZAR DEMANDA
df_demanda_agg.cache()
df_demanda_agg.count()
print("Demanda materializada en cache")

# ===============================================================================
# 4. PREPARACIÓN DE FACTURACIÓN (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO DATOS DE FACTURACIÓN ===")

df_facturacion_prep = df_facturacion.filter(
    col("Material").isin(skus_validos)
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"), 
    col("Volumen_ventas").alias("Facturacion"),
    col("Ejercicio_Periodo"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero")
).filter(
    (col("Ano") >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
)

df_facturacion_agg = df_facturacion_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Facturacion").alias("Facturacion_total")
)

print(f"Facturación agregada: {df_facturacion_agg.count():,} registros")

# MATERIALIZAR FACTURACIÓN
df_facturacion_agg.cache()
df_facturacion_agg.count()
print("Facturación materializada en cache")

# ===============================================================================
# 5. PREPARACIÓN DE LLEGADAS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO LLEGADAS ===")

# Entregas
df_entregas_prep = df_entregas.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_entrega_confirmada")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_entrega_confirmada")).alias("Ano"),
    month(col("Fecha_entrega_confirmada")).alias("mes_numero"),
    col("Cantidad_Entregada_Efectivamente").alias("Entregas")
)

df_entregas_agg = df_entregas_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Entregas").alias("Entregas_total")
)

# Compras
df_compras_prep = df_compras.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_Final")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_Final")).alias("Ano"),
    month(col("Fecha_Final")).alias("mes_numero"),
    col("Qty_sin_EE").alias("Compras")
)

df_compras_agg = df_compras_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Compras").alias("Compras_total")
)

# Combinar entregas + compras = llegadas
df_entregas_union = df_entregas_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Entregas_total").alias("Cantidad")
)

df_compras_union = df_compras_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Compras_total").alias("Cantidad")
)

df_llegadas_combinado = df_entregas_union.union(df_compras_union)

df_llegadas_agg = df_llegadas_combinado.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Cantidad").alias("Llegadas_total")
)

print(f"Llegadas agregadas: {df_llegadas_agg.count():,} registros")

# MATERIALIZAR LLEGADAS
df_llegadas_agg.cache()
df_llegadas_agg.count()
print("Llegadas materializadas en cache")

# ===============================================================================
# 6. PROCESAMIENTO DE INVENTARIOS POR SEMANA - CORREGIDO CON TRANSICIÓN DE AÑO
# ===============================================================================

print("=== PROCESANDO INVENTARIOS POR SEMANA - MÉTODO CORREGIDO CON TRANSICIÓN DE AÑO ===")

# PASO 1: AGREGAR por País+Familia+Semana - INCLUYE AÑO ANTERIOR PARA CONTINUIDAD
df_inventarios_por_semana = df_costeado.filter(
    (col("Sociedad_Pais").isin(paises_requeridos)) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Grupo_Articulos") != "-") &
    # ✅ CORRECCIÓN: Incluye año anterior para transición de año
    (substring(col("Ejercicio___Periodo"), 1, 4).cast("int") >= ano_actual - 1)
).groupBy(
    col("Sociedad_Pais").alias("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    substring(col("Ejercicio___Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio___Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Semana")
).agg(
    F.sum("Cantidad_Stock_Total_").alias("Inventario_semana")
)

print(f"✅ INCLUYE AÑO ANTERIOR ({ano_actual-1}) para continuidad en transición de año")

# MATERIALIZAR INVENTARIOS
df_inventarios_por_semana.cache()
df_inventarios_por_semana.count()
print("Inventarios por semana materializados")

# PASO 2: NUEVA LÓGICA - Inventario Final por Mes (última semana del mes)
df_limites_semanas = df_inventarios_por_semana.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    spark_max("Semana").alias("ultima_semana")
)

# Inventario final = última semana del mes actual
df_inv_final = df_inventarios_por_semana.alias("inv").join(
    df_limites_semanas.alias("lim"),
    (col("inv.Pais") == col("lim.Pais")) &
    (col("inv.Familia") == col("lim.Familia")) &
    (col("inv.Ano") == col("lim.Ano")) &
    (col("inv.mes_numero") == col("lim.mes_numero")) &
    (col("inv.Semana") == col("lim.ultima_semana")),
    "inner"
).select(
    col("inv.Pais"), 
    col("inv.Familia"), 
    col("inv.Ano"), 
    col("inv.mes_numero"),
    col("inv.Inventario_semana").alias("Inv_Final_Mes")
)

print(f"Inventarios finales por mes calculados: {df_inv_final.count():,} registros")

# PASO 3: NUEVA LÓGICA - Inventario Inicial = Inventario Final del mes anterior
# Usar window function para obtener el inventario final del mes anterior

# ✅ CORRECCIÓN: Window ordenado por Año y Mes - SIN FILTROS para permitir transición de año
windowSpec = Window.partitionBy("Pais", "Familia").orderBy("Ano", "mes_numero")

# Aplicar LAG para obtener el inventario final del mes anterior (incluso del año anterior)
df_inventarios_con_inicial_completo = df_inv_final.withColumn(
    "Inv_Inicial_Mes", 
    lag(col("Inv_Final_Mes"), 1).over(windowSpec)
).withColumn(
    # Si es NULL (primer mes disponible), poner 0
    "Inv_Inicial", 
    when(col("Inv_Inicial_Mes").isNull(), lit(0.0))
    .otherwise(col("Inv_Inicial_Mes"))
).select(
    "Pais", "Familia", "Ano", "mes_numero",
    "Inv_Inicial",
    col("Inv_Final_Mes").alias("Inv_Final")
)

# ✅ AHORA SÍ FILTRAR: Solo mantener datos del año actual en la salida final
df_inventarios_con_inicial = df_inventarios_con_inicial_completo.filter(
    col("Ano") >= ano_actual
)

print(f"Inventarios iniciales calculados (con LAG y transición de año): {df_inventarios_con_inicial.count():,} registros")

# MATERIALIZAR INVENTARIOS MENSUALES CORREGIDOS
df_inventarios_con_inicial.cache()
df_inventarios_con_inicial.count()
print("Inventarios mensuales CORREGIDOS materializados")

# VERIFICACIÓN de la corrección CON TRANSICIÓN DE AÑO
print("\n=== VERIFICACIÓN DE CORRECCIÓN EN INVENTARIOS CON TRANSICIÓN DE AÑO ===")
print("Muestra incluyendo transición Diciembre → Enero:")

# Mostrar últimos meses del año anterior + primeros del actual (si disponible)
df_inventarios_con_inicial_completo.filter(
    (col("Pais") == paises_requeridos[0]) &
    (col("Familia").isNotNull()) &
    (((col("Ano") == ano_actual - 1) & (col("mes_numero") >= 11)) |  # Nov-Dic año anterior
     ((col("Ano") == ano_actual) & (col("mes_numero") <= 3)))        # Ene-Mar año actual
).orderBy("Ano", "mes_numero").select(
    "Ano", "mes_numero", "Inv_Inicial", "Inv_Final"
).show(10, truncate=False)

print("✅ TRANSICIÓN DE AÑO: Inv_Inicial Enero = Inv_Final Diciembre año anterior")

# ===============================================================================
# 7. CREAR ESTRUCTURA BASE Y JOINS (ACTUALIZADO)
# ===============================================================================

print("=== CREANDO ESTRUCTURA BASE ===")

# Obtener combinaciones únicas de País + Familia desde inventarios reales
paises_familias_inv = df_inventarios_con_inicial.select("Pais", "Familia").distinct()

# Estructura base con inventarios CORREGIDOS
df_inventarios_master = paises_familias_inv \
    .crossJoin(df_anos) \
    .crossJoin(df_meses) \
    .alias("base") \
    .join(
        df_inventarios_con_inicial.alias("inv"),
        (col("base.Pais") == col("inv.Pais")) &
        (col("base.Familia") == col("inv.Familia")) &
        (col("base.Ano") == col("inv.Ano")) &
        (col("base.ORDENM") == col("inv.mes_numero")),
        "left"
    ).select(
        col("base.Ano"),
        lit(mes_ejecucion_actual).alias("Mes"),
        col("base.Pais").alias("PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        lit(1).alias("PWOH"),
        F.coalesce(col("inv.Inv_Inicial"), F.lit(0.0)).alias("Inv_Inicial"),
        F.coalesce(col("inv.Inv_Final"), F.lit(0.0)).alias("Inv_Final")
    )

print(f"Estructura base con inventarios CORREGIDOS: {df_inventarios_master.count():,} registros")

# MATERIALIZAR ESTRUCTURA BASE
df_inventarios_master.cache()
df_inventarios_master.count()
print("Estructura base materializada")

# ===============================================================================
# 8. JOINS CON TODAS LAS MÉTRICAS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== UNIENDO TODAS LAS MÉTRICAS ===")

# JOIN con demanda
df_con_demanda = df_inventarios_master.alias("base") \
    .join(
        df_demanda_agg.alias("dem"),
        (col("base.PAIS") == col("dem.Pais")) &
        (col("base.Familia") == col("dem.Familia")) &
        (col("base.Ano") == col("dem.Ano")) &
        (col("base.ORDENM") == col("dem.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("dem.Fcst_total"), lit(0.0)).alias("Fcst")
    )

# JOIN con facturación
df_con_facturacion = df_con_demanda.alias("base") \
    .join(
        df_facturacion_agg.alias("fact"),
        (col("base.PAIS") == col("fact.Pais")) &
        (col("base.Familia") == col("fact.Familia")) &
        (col("base.Ano") == col("fact.Ano")) &
        (col("base.ORDENM") == col("fact.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("fact.Facturacion_total"), lit(0.0)).alias("Facturacion")
    )

# JOIN con llegadas
df_master_basico = df_con_facturacion.alias("base") \
    .join(
        df_llegadas_agg.alias("lleg"),
        (col("base.PAIS") == col("lleg.Pais")) &
        (col("base.Familia") == col("lleg.Familia")) &
        (col("base.Ano") == col("lleg.Ano")) &
        (col("base.ORDENM") == col("lleg.mes_numero")),
        "left"
    ).select(
        # Dimensiones base
        col("base.Ano"),
        col("base.Mes"),
        col("base.PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        col("base.PWOH"),
        
        # Todas las métricas básicas
        col("base.Inv_Inicial"),
        col("base.Inv_Final"),
        coalesce(col("lleg.Llegadas_total"), lit(0.0)).alias("Llegadas"),
        col("base.Fcst"),
        col("base.Facturacion")
    )

print(f"Tabla maestra básica: {df_master_basico.count():,} registros")

# ===============================================================================
# 9. GUARDAR TABLA BÁSICA EN LAKEHOUSE
# ===============================================================================

print("=== GUARDANDO TABLA BÁSICA CORREGIDA ===")

# Limpiar cache antes de guardar
spark.catalog.clearCache()

# Optimizar particionado
df_master_basico_optimizado = df_master_basico.repartition(100, "PAIS", "Familia")

# Guardar tabla básica para el notebook de cálculos
spark.sql("USE LH_INTELLIGENT_VISIBILITY_CARATULAS")

try:
    spark.sql("DROP TABLE IF EXISTS continental_master_basic_daily")
    print("Tabla anterior eliminada")
except:
    print("No había tabla anterior")

df_master_basico_optimizado.write.format("delta").mode("overwrite").saveAsTable("continental_master_basic_daily")

print("TABLA BÁSICA CORREGIDA GUARDADA:")
print("- Nombre: continental_master_basic_daily")
print("- Ubicación: LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_master_basic_daily")
print("- Contiene: Inv_Inicial (MES ANTERIOR), Inv_Final (MES ACTUAL), Llegadas, Fcst, Facturación")
print("- CORRECCIÓN APLICADA: Inv_Inicial = última semana del mes anterior")
print("- Optimizada para cálculos avanzados")

# ===============================================================================
# 10. VERIFICACIÓN FINAL
# ===============================================================================

print("=== VERIFICACIÓN FINAL ETL DIARIO CORREGIDO ===")

print("Sample CR-LAVA con inventarios CORREGIDOS:")
df_master_basico.filter(
    (col("PAIS") == "CR") & (col("Familia") == "LAVA") & (col("Ano") == ano_actual)
).select("Attribute", "ORDENM", "Inv_Inicial", "Inv_Final", "Llegadas", "Fcst", "Facturacion") \
.orderBy("ORDENM").show(12, truncate=False)

print("\n=== VALIDACIÓN DE CORRECCIÓN CON TRANSICIÓN DE AÑO ===")
print("Verificando continuidad Diciembre → Enero:")

# Mostrar transición de año si hay datos disponibles
df_validacion_transicion = df_master_basico.filter(
    (col("PAIS") == "CO") & (col("Familia") == "LAVA") &
    (((col("Ano") == ano_actual - 1) & (col("ORDENM") >= 12)) |  # Nov-Dic año anterior  
     ((col("Ano") == ano_actual) & (col("ORDENM") <= 12)))        # Ene-Mar año actual
).orderBy("Ano", "ORDENM").select("Ano", "ORDENM", "Attribute", "Inv_Inicial", "Inv_Final")

df_validacion_transicion.show(10, truncate=False)

print("ETL DIARIO CORREGIDO COMPLETADO EXITOSAMENTE")
print("✅ CORRECCIÓN APLICADA: Inventarios iniciales ahora toman última semana del mes anterior")
print("✅ TRANSICIÓN DE AÑO: Maneja correctamente el paso de Diciembre → Enero")
print("✅ CONTINUIDAD GARANTIZADA: Inv_Inicial Enero = Inv_Final Diciembre año anterior")
print("LISTO PARA NOTEBOOK DE CÁLCULOS AVANZADOS")
```

```python
#!/usr/bin/env python
# coding: utf-8

# ## NB_ETL_DAILY_CARATULAS - VERSIÓN FINAL CORREGIDA
# 
# Corrección fundamental en inventarios:
# - Inv_Inicial: Última semana del mes ANTERIOR
# - Inv_Final: Última semana del mes ACTUAL
# 
# Complemento temporal CSV 2026:
# - Prioridad a cubos SOLO si tiene valor > 0
# - Si cubos = 0 o NULL → usa CSV
# - Transición automática cuando cubos se actualicen

# ===============================================================================
# NOTEBOOK 1: CONTINENTAL_MASTER_ETL_DAILY - VERSIÓN FINAL
# Actualización diaria - Solo procesamiento básico y estable
# ===============================================================================

from datetime import datetime
from pyspark.sql.functions import (
    col, year, month, max as spark_max, min as spark_min, count, lit, 
    substring, coalesce, sum as spark_sum, lag, when, isnan, isnull, countDistinct
)
from pyspark.sql.types import *
import pyspark.sql.functions as F
from pyspark.sql.window import Window

print("=== CONTINENTAL MASTER ETL DAILY - VERSIÓN FINAL CORREGIDA ===")

# ===============================================================================
# 1. CONFIGURACIÓN GLOBAL
# ===============================================================================

# dict_skus ya debe estar disponible desde la celda anterior

# PARAMETROS GLOBALES
fecha_actual = datetime.now()
ano_actual = fecha_actual.year
mes_actual_num = fecha_actual.month

# Mapeo de meses
nombres_meses = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril", 5: "Mayo", 6: "Junio",
    7: "Julio", 8: "Agosto", 9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

meses_mapping = {
    1: ("ENE", "Enero"), 2: ("FEB", "Febrero"), 3: ("MAR", "Marzo"), 
    4: ("ABR", "Abril"), 5: ("MAY", "Mayo"), 6: ("JUN", "Junio"),
    7: ("JUL", "Julio"), 8: ("AGO", "Agosto"), 9: ("SEP", "Septiembre"), 
    10: ("OCT", "Octubre"), 11: ("NOV", "Noviembre"), 12: ("DIC", "Diciembre")
}

mes_ejecucion_actual = nombres_meses[mes_actual_num]
anos_a_procesar = list(range(ano_actual, ano_actual + 6))

# Extraer SKUs válidos
skus_validos = []
for region, paises in dict_skus.items():
    for pais, familias in paises.items():
        for familia, skus in familias.items():
            skus_validos.extend(skus)

paises_requeridos = ["CR","GT","HN","SV","PE","CO","EC","CL","AR","CA","MX"]

print(f"Fecha actual del sistema: {fecha_actual.strftime('%Y-%m-%d')}")
print(f"Año actual: {ano_actual}")
print(f"Mes de ejecución: {mes_actual_num} ({mes_ejecucion_actual})")
print(f"Años a procesar: {anos_a_procesar}")
print(f"Total SKUs válidos: {len(skus_validos)}")
print(f"Países requeridos: {paises_requeridos}")

# ===============================================================================
# 2. CARGA DE DATOS
# ===============================================================================

print("\n=== CARGANDO TABLAS DEL LAKEHOUSE ===")

df_demanda = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_demanda")
df_entregas = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_entregas") 
df_costeado = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_costeado")
df_calendario = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_calendario")
df_compras = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_compras")
df_facturacion = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_facturacion")

print("Tablas cargadas exitosamente")
print(f"  - continental_demanda: {df_demanda.count():,} registros")
print(f"  - continental_entregas: {df_entregas.count():,} registros")
print(f"  - continental_costeado: {df_costeado.count():,} registros")
print(f"  - continental_compras: {df_compras.count():,} registros")
print(f"  - continental_facturacion: {df_facturacion.count():,} registros")

# ===============================================================================
# 3. PREPARACIÓN DE DEMANDA CON COMPLEMENTO CSV 2026
# ===============================================================================

print("\n" + "="*80)
print("=== PREPARANDO DATOS DE DEMANDA CON COMPLEMENTO CSV 2026 ===")
print("="*80)

# Crear DataFrame de meses para cross join
meses_data = [(mes_num, codigo, nombre, mes_num) for mes_num, (codigo, nombre) in meses_mapping.items()]
df_meses = spark.createDataFrame(meses_data, ["ORDENM", "Attribute", "Mes_nombre", "mes_numero"])
df_anos = spark.createDataFrame([(ano,) for ano in anos_a_procesar], ["Ano"])

# -------------------------------------------------------------------------------
# PASO 1: CARGAR DEMANDA DE CUBOS (FUENTE PRINCIPAL)
# -------------------------------------------------------------------------------

print("\n→ PASO 1: Cargando demanda desde CUBOS (fuente principal)...")

# Filtrar demanda solo con SKUs válidos desde año actual
df_demanda_prep = df_demanda.filter(
    (substring(col("Ejercicio_Periodo"), 1, 4).cast("int") >= ano_actual) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Pais").isNotNull()) &
    (col("Material").isin(skus_validos)) &
    (col("Grupo_Articulos") != "-")
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
)

# Agrupar demanda por dimensiones principales
df_demanda_cubos = df_demanda_prep.filter(
    col("Pais").isin(paises_requeridos)
).groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Fcst").alias("Fcst_cubos"),
    F.count("Material").alias("num_materiales_cubos")
)

print(f"  ✓ Demanda desde CUBOS: {df_demanda_cubos.count():,} registros")

# Verificar cobertura de cubos en 2026
print("\n  📊 Cobertura de CUBOS en 2026:")
df_cobertura_cubos_2026 = df_demanda_cubos.filter(col("Ano") == 2026).groupBy("Ano").agg(
    countDistinct("mes_numero").alias("Meses_con_datos"),
    countDistinct("Pais").alias("Paises"),
    spark_sum("Fcst_cubos").alias("Fcst_Total")
)
df_cobertura_cubos_2026.show()

# Detectar meses con valores reales (> 0) vs meses con 0
print("\n  📊 Detalle de CUBOS 2026 - Meses con valores REALES (> 0):")
df_demanda_cubos.filter(col("Ano") == 2026).groupBy("mes_numero").agg(
    spark_sum(when(col("Fcst_cubos") > 0, 1).otherwise(0)).alias("Registros_con_valor"),
    spark_sum(when(col("Fcst_cubos") == 0, 1).otherwise(0)).alias("Registros_en_cero"),
    spark_sum("Fcst_cubos").alias("Fcst_Total")
).orderBy("mes_numero").show(12)

meses_cubos_2026 = df_cobertura_cubos_2026.first()
if meses_cubos_2026:
    meses_con_datos = meses_cubos_2026["Meses_con_datos"]
    print(f"  → Cubos tienen datos para {meses_con_datos}/12 meses de 2026")
    if meses_con_datos < 12:
        print(f"  ⚠️  Faltan {12 - meses_con_datos} meses → Se complementará con CSV")
else:
    print("  ⚠️  No hay datos de cubos para 2026")

# -------------------------------------------------------------------------------
# PASO 2: CARGAR CSV TEMPORAL 2026 (COMPLEMENTO)
# -------------------------------------------------------------------------------

csv_cargado = False

print("\n→ PASO 2: Intentando cargar CSV temporal 2026 (complemento)...")

try:
    # ✅ CARGAR CSV desde Files
    df_csv_2026_raw = spark.read.format("csv") \
        .option("header", "true") \
        .option("inferSchema", "true") \
        .load("Files/BGT_2026.csv")
    
    print("  ✓ CSV cargado desde Files/BGT_2026.csv")
    csv_cargado = True
    
    # Mostrar info del CSV
    print(f"  → Columnas del CSV: {df_csv_2026_raw.columns}")
    print(f"  → Registros crudos: {df_csv_2026_raw.count():,}")
    
    # ✅ Procesar CSV con columnas correctas (con acento)
    df_csv_prep = df_csv_2026_raw.filter(
        (col("País").isin(paises_requeridos)) &
        (col("Ejercicio_Periodo").isNotNull()) &
        (col("Grupo_Articulos").isNotNull())
    ).select(
        col("País").alias("Pais"),
        col("Grupo_Articulos").alias("Familia"),
        col("Ejercicio_Periodo"),
        col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
    )
    
    print(f"  → Registros después de filtros de país: {df_csv_prep.count():,}")
    
    # Extraer año y mes del Ejercicio_Periodo
    df_csv_prep = df_csv_prep.withColumn(
        "Ano", 
        (col("Ejercicio_Periodo") / 1000).cast("int")
    ).withColumn(
        "mes_numero", 
        (col("Ejercicio_Periodo") % 1000).cast("int")
    )
    
    # Filtrar solo 2026
    df_csv_prep = df_csv_prep.filter(
        (col("Ano") == 2026) &
        (col("mes_numero").between(1, 12))
    )
    
    print(f"  → Registros filtrados para 2026: {df_csv_prep.count():,}")
    
    # Agregar CSV
    df_demanda_csv = df_csv_prep.groupBy(
        "Pais", "Familia", "Ano", "mes_numero"
    ).agg(
        F.sum("Fcst").alias("Fcst_csv"),
        F.count("*").alias("num_registros_csv")
    )
    
    print(f"  ✓ Demanda desde CSV 2026 agregada: {df_demanda_csv.count():,} registros")
    
    # Verificar cobertura del CSV
    print("\n  📊 Cobertura del CSV 2026:")
    df_demanda_csv.groupBy("Ano").agg(
        countDistinct("mes_numero").alias("Meses_con_datos"),
        countDistinct("Pais").alias("Paises"),
        spark_sum("Fcst_csv").alias("Fcst_Total")
    ).show()
    
    # Mostrar detalle por mes
    print("\n  📊 Detalle del CSV por mes:")
    df_demanda_csv.groupBy("mes_numero").agg(
        countDistinct("Pais").alias("Paises"),
        countDistinct("Familia").alias("Familias"),
        spark_sum("Fcst_csv").alias("Fcst_Total")
    ).orderBy("mes_numero").show(12)
    
except Exception as e:
    print(f"\n  ⚠️  Error al cargar CSV: {str(e)}")
    print(f"  → Tipo de error: {type(e).__name__}")
    csv_cargado = False

# -------------------------------------------------------------------------------
# PASO 3: COMBINAR CUBOS + CSV (PRIORIDAD A CUBOS CON VALOR REAL > 0)
# -------------------------------------------------------------------------------

if csv_cargado:
    print("\n→ PASO 3: Combinando CUBOS + CSV (prioridad a CUBOS con valores REALES)...")
    print("  🔑 LÓGICA CORREGIDA: Si cubos > 0 → usar CUBOS, si cubos = 0 o NULL → usar CSV")
    
    # FULL OUTER JOIN para tener todos los registros
    df_demanda_completa = df_demanda_cubos.alias("cubos").join(
        df_demanda_csv.alias("csv"),
        ["Pais", "Familia", "Ano", "mes_numero"],
        "full_outer"
    ).select(
        coalesce(col("cubos.Pais"), col("csv.Pais")).alias("Pais"),
        coalesce(col("cubos.Familia"), col("csv.Familia")).alias("Familia"),
        coalesce(col("cubos.Ano"), col("csv.Ano")).alias("Ano"),
        coalesce(col("cubos.mes_numero"), col("csv.mes_numero")).alias("mes_numero"),
        
        # ✅ CORRECCIÓN CRÍTICA: Solo usar cubos si tiene VALOR REAL (> 0)
        when(
            (col("cubos.Fcst_cubos").isNotNull()) & (col("cubos.Fcst_cubos") > 0),
            col("cubos.Fcst_cubos")
        ).otherwise(
            coalesce(col("csv.Fcst_csv"), lit(0.0))
        ).alias("Fcst_total"),
        
        # Columnas de diagnóstico
        col("cubos.Fcst_cubos").alias("_debug_fcst_cubos"),
        col("csv.Fcst_csv").alias("_debug_fcst_csv"),
        
        # Indicador de fuente corregido
        when(
            (col("cubos.Fcst_cubos").isNotNull()) & (col("cubos.Fcst_cubos") > 0), 
            lit("CUBOS")
        ).when(
            col("csv.Fcst_csv").isNotNull(), 
            lit("CSV")
        ).otherwise(
            lit("SIN_DATO")
        ).alias("_debug_fuente")
    )
    
    print(f"  ✓ Demanda TOTAL combinada: {df_demanda_completa.count():,} registros")
    
    # Verificar cobertura por año
    print("\n  📊 COBERTURA FINAL POR AÑO:")
    df_demanda_completa.groupBy("Ano").agg(
        spark_sum("Fcst_total").alias("Fcst_Total"),
        countDistinct("mes_numero").alias("Meses_con_datos"),
        countDistinct("Pais").alias("Paises")
    ).orderBy("Ano").show()
    
    # Verificar qué meses de 2026 vienen de cada fuente
    print("\n  📊 DETALLE DE FUENTES PARA 2026 (por mes) - CORREGIDO:")
    df_verificacion_2026 = df_demanda_completa.filter(col("Ano") == 2026).groupBy("mes_numero").agg(
        spark_sum(when(col("_debug_fuente") == "CUBOS", 1).otherwise(0)).alias("Reg_CUBOS"),
        spark_sum(when(col("_debug_fuente") == "CSV", 1).otherwise(0)).alias("Reg_CSV"),
        spark_sum(when(col("_debug_fuente") == "SIN_DATO", 1).otherwise(0)).alias("Sin_Dato"),
        spark_sum("Fcst_total").alias("Fcst_Total")
    ).orderBy("mes_numero")
    
    df_verificacion_2026.show(12)
    
    # Verificación específica de mayo (el mes problemático)
    print("\n  🔍 VERIFICACIÓN ESPECIAL: MAYO 2026 (mes 5) - debe usar CSV:")
    df_demanda_completa.filter(
        (col("Ano") == 2026) & (col("mes_numero") == 5)
    ).groupBy("_debug_fuente").agg(
        count("*").alias("Registros"),
        spark_sum("Fcst_total").alias("Fcst_Total")
    ).show()
    
    # Mostrar resumen por país para 2026
    print("\n  📊 RESUMEN POR PAÍS PARA 2026:")
    df_demanda_completa.filter(col("Ano") == 2026).groupBy("Pais").agg(
        countDistinct("mes_numero").alias("Meses"),
        spark_sum(when(col("_debug_fuente") == "CUBOS", 1).otherwise(0)).alias("Reg_CUBOS"),
        spark_sum(when(col("_debug_fuente") == "CSV", 1).otherwise(0)).alias("Reg_CSV"),
        spark_sum("Fcst_total").alias("Fcst_Total")
    ).orderBy("Pais").show(15)
    
    # Usar la demanda completa para el resto del proceso
    df_demanda_agg = df_demanda_completa.select(
        "Pais", "Familia", "Ano", "mes_numero", "Fcst_total"
    )
    
    print("\n  ✅ Demanda combinada CUBOS+CSV lista para procesamiento")
    print("  ✅ CORRECCIÓN APLICADA: Cubos = 0 ahora usan CSV correctamente")
    
else:
    print("\n→ PASO 3: Usando solo datos de CUBOS (CSV no disponible)...")
    
    # Si no hay CSV, usar solo cubos
    df_demanda_agg = df_demanda_cubos.withColumnRenamed("Fcst_cubos", "Fcst_total")
    
    print("  ✅ Demanda desde CUBOS únicamente")

# -------------------------------------------------------------------------------
# RESUMEN FINAL DE DEMANDA
# -------------------------------------------------------------------------------

print(f"\n{'='*80}")
print(f"DEMANDA FINAL PREPARADA: {df_demanda_agg.count():,} registros")
print(f"{'='*80}")

# MATERIALIZAR DEMANDA
df_demanda_agg.cache()
df_demanda_agg.count()
print("✓ Demanda materializada en cache")

# ===============================================================================
# 4. PREPARACIÓN DE FACTURACIÓN
# ===============================================================================

print("\n=== PREPARANDO DATOS DE FACTURACIÓN ===")

df_facturacion_prep = df_facturacion.filter(
    col("Material").isin(skus_validos)
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"), 
    col("Volumen_ventas").alias("Facturacion"),
    col("Ejercicio_Periodo"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero")
).filter(
    (col("Ano") >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
)

df_facturacion_agg = df_facturacion_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Facturacion").alias("Facturacion_total")
)

print(f"Facturación agregada: {df_facturacion_agg.count():,} registros")

# MATERIALIZAR FACTURACIÓN
df_facturacion_agg.cache()
df_facturacion_agg.count()
print("Facturación materializada en cache")

# ===============================================================================
# 5. PREPARACIÓN DE LLEGADAS
# ===============================================================================

print("\n=== PREPARANDO LLEGADAS ===")

# Entregas
df_entregas_prep = df_entregas.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_entrega_confirmada")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_entrega_confirmada")).alias("Ano"),
    month(col("Fecha_entrega_confirmada")).alias("mes_numero"),
    col("Cantidad_Entregada_Efectivamente").alias("Entregas")
)

df_entregas_agg = df_entregas_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Entregas").alias("Entregas_total")
)

# Compras
df_compras_prep = df_compras.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_Final")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_Final")).alias("Ano"),
    month(col("Fecha_Final")).alias("mes_numero"),
    col("Qty_sin_EE").alias("Compras")
)

df_compras_agg = df_compras_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Compras").alias("Compras_total")
)

# Combinar entregas + compras = llegadas
df_entregas_union = df_entregas_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Entregas_total").alias("Cantidad")
)

df_compras_union = df_compras_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Compras_total").alias("Cantidad")
)

df_llegadas_combinado = df_entregas_union.union(df_compras_union)

df_llegadas_agg = df_llegadas_combinado.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Cantidad").alias("Llegadas_total")
)

print(f"Llegadas agregadas: {df_llegadas_agg.count():,} registros")

# MATERIALIZAR LLEGADAS
df_llegadas_agg.cache()
df_llegadas_agg.count()
print("Llegadas materializadas en cache")

# ===============================================================================
# 6. PROCESAMIENTO DE INVENTARIOS POR SEMANA - CORREGIDO CON TRANSICIÓN DE AÑO
# ===============================================================================

print("\n=== PROCESAMIENTO DE INVENTARIOS - MÉTODO CORREGIDO CON TRANSICIÓN DE AÑO ===")

# PASO 1: AGREGAR por País+Familia+Semana - INCLUYE AÑO ANTERIOR PARA CONTINUIDAD
df_inventarios_por_semana = df_costeado.filter(
    (col("Sociedad_Pais").isin(paises_requeridos)) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Grupo_Articulos") != "-") &
    # ✅ CORRECCIÓN: Incluye año anterior para transición de año
    (substring(col("Ejercicio___Periodo"), 1, 4).cast("int") >= ano_actual - 1)
).groupBy(
    col("Sociedad_Pais").alias("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    substring(col("Ejercicio___Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio___Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Semana")
).agg(
    F.sum("Cantidad_Stock_Total_").alias("Inventario_semana")
)

print(f"✅ INCLUYE AÑO ANTERIOR ({ano_actual-1}) para continuidad en transición de año")

# MATERIALIZAR INVENTARIOS
df_inventarios_por_semana.cache()
df_inventarios_por_semana.count()
print("Inventarios por semana materializados")

# PASO 2: Inventario Final por Mes (última semana del mes)
df_limites_semanas = df_inventarios_por_semana.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    spark_max("Semana").alias("ultima_semana")
)

# Inventario final = última semana del mes actual
df_inv_final = df_inventarios_por_semana.alias("inv").join(
    df_limites_semanas.alias("lim"),
    (col("inv.Pais") == col("lim.Pais")) &
    (col("inv.Familia") == col("lim.Familia")) &
    (col("inv.Ano") == col("lim.Ano")) &
    (col("inv.mes_numero") == col("lim.mes_numero")) &
    (col("inv.Semana") == col("lim.ultima_semana")),
    "inner"
).select(
    col("inv.Pais"), 
    col("inv.Familia"), 
    col("inv.Ano"), 
    col("inv.mes_numero"),
    col("inv.Inventario_semana").alias("Inv_Final_Mes")
)

print(f"Inventarios finales por mes calculados: {df_inv_final.count():,} registros")

# PASO 3: Inventario Inicial = Inventario Final del mes anterior
# Usar window function para obtener el inventario final del mes anterior

# ✅ CORRECCIÓN: Window ordenado por Año y Mes - SIN FILTROS para permitir transición de año
windowSpec = Window.partitionBy("Pais", "Familia").orderBy("Ano", "mes_numero")

# Aplicar LAG para obtener el inventario final del mes anterior (incluso del año anterior)
df_inventarios_con_inicial_completo = df_inv_final.withColumn(
    "Inv_Inicial_Mes", 
    lag(col("Inv_Final_Mes"), 1).over(windowSpec)
).withColumn(
    # Si es NULL (primer mes disponible), poner 0
    "Inv_Inicial", 
    when(col("Inv_Inicial_Mes").isNull(), lit(0.0))
    .otherwise(col("Inv_Inicial_Mes"))
).select(
    "Pais", "Familia", "Ano", "mes_numero",
    "Inv_Inicial",
    col("Inv_Final_Mes").alias("Inv_Final")
)

# ✅ AHORA SÍ FILTRAR: Solo mantener datos del año actual en la salida final
df_inventarios_con_inicial = df_inventarios_con_inicial_completo.filter(
    col("Ano") >= ano_actual
)

print(f"Inventarios iniciales calculados (con LAG y transición de año): {df_inventarios_con_inicial.count():,} registros")

# MATERIALIZAR INVENTARIOS MENSUALES CORREGIDOS
df_inventarios_con_inicial.cache()
df_inventarios_con_inicial.count()
print("Inventarios mensuales CORREGIDOS materializados")

# VERIFICACIÓN de la corrección CON TRANSICIÓN DE AÑO
print("\n=== VERIFICACIÓN DE CORRECCIÓN EN INVENTARIOS CON TRANSICIÓN DE AÑO ===")
print("Muestra incluyendo transición Diciembre → Enero:")

# Mostrar últimos meses del año anterior + primeros del actual (si disponible)
df_inventarios_con_inicial_completo.filter(
    (col("Pais") == paises_requeridos[0]) &
    (col("Familia").isNotNull()) &
    (((col("Ano") == ano_actual - 1) & (col("mes_numero") >= 11)) |  # Nov-Dic año anterior
     ((col("Ano") == ano_actual) & (col("mes_numero") <= 3)))        # Ene-Mar año actual
).orderBy("Ano", "mes_numero").select(
    "Ano", "mes_numero", "Inv_Inicial", "Inv_Final"
).show(10, truncate=False)

print("✅ TRANSICIÓN DE AÑO: Inv_Inicial Enero = Inv_Final Diciembre año anterior")

# ===============================================================================
# 7. CREAR ESTRUCTURA BASE Y JOINS
# ===============================================================================

print("\n=== CREANDO ESTRUCTURA BASE ===")

# Obtener combinaciones únicas de País + Familia desde inventarios reales
paises_familias_inv = df_inventarios_con_inicial.select("Pais", "Familia").distinct()

# Estructura base con inventarios CORREGIDOS
df_inventarios_master = paises_familias_inv \
    .crossJoin(df_anos) \
    .crossJoin(df_meses) \
    .alias("base") \
    .join(
        df_inventarios_con_inicial.alias("inv"),
        (col("base.Pais") == col("inv.Pais")) &
        (col("base.Familia") == col("inv.Familia")) &
        (col("base.Ano") == col("inv.Ano")) &
        (col("base.ORDENM") == col("inv.mes_numero")),
        "left"
    ).select(
        col("base.Ano"),
        lit(mes_ejecucion_actual).alias("Mes"),
        col("base.Pais").alias("PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        lit(1).alias("PWOH"),
        F.coalesce(col("inv.Inv_Inicial"), F.lit(0.0)).alias("Inv_Inicial"),
        F.coalesce(col("inv.Inv_Final"), F.lit(0.0)).alias("Inv_Final")
    )

print(f"Estructura base con inventarios CORREGIDOS: {df_inventarios_master.count():,} registros")

# MATERIALIZAR ESTRUCTURA BASE
df_inventarios_master.cache()
df_inventarios_master.count()
print("Estructura base materializada")

# ===============================================================================
# 8. JOINS CON TODAS LAS MÉTRICAS
# ===============================================================================

print("\n=== UNIENDO TODAS LAS MÉTRICAS ===")

# JOIN con demanda
df_con_demanda = df_inventarios_master.alias("base") \
    .join(
        df_demanda_agg.alias("dem"),
        (col("base.PAIS") == col("dem.Pais")) &
        (col("base.Familia") == col("dem.Familia")) &
        (col("base.Ano") == col("dem.Ano")) &
        (col("base.ORDENM") == col("dem.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("dem.Fcst_total"), lit(0.0)).alias("Fcst")
    )

# JOIN con facturación
df_con_facturacion = df_con_demanda.alias("base") \
    .join(
        df_facturacion_agg.alias("fact"),
        (col("base.PAIS") == col("fact.Pais")) &
        (col("base.Familia") == col("fact.Familia")) &
        (col("base.Ano") == col("fact.Ano")) &
        (col("base.ORDENM") == col("fact.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("fact.Facturacion_total"), lit(0.0)).alias("Facturacion")
    )

# JOIN con llegadas
df_master_basico = df_con_facturacion.alias("base") \
    .join(
        df_llegadas_agg.alias("lleg"),
        (col("base.PAIS") == col("lleg.Pais")) &
        (col("base.Familia") == col("lleg.Familia")) &
        (col("base.Ano") == col("lleg.Ano")) &
        (col("base.ORDENM") == col("lleg.mes_numero")),
        "left"
    ).select(
        # Dimensiones base
        col("base.Ano"),
        col("base.Mes"),
        col("base.PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        col("base.PWOH"),
        
        # Todas las métricas básicas
        col("base.Inv_Inicial"),
        col("base.Inv_Final"),
        coalesce(col("lleg.Llegadas_total"), lit(0.0)).alias("Llegadas"),
        col("base.Fcst"),
        col("base.Facturacion")
    )

print(f"Tabla maestra básica: {df_master_basico.count():,} registros")

# ===============================================================================
# 9. GUARDAR TABLA BÁSICA EN LAKEHOUSE
# ===============================================================================

print("\n=== GUARDANDO TABLA BÁSICA CORREGIDA ===")

# Limpiar cache antes de guardar
spark.catalog.clearCache()

# Optimizar particionado
df_master_basico_optimizado = df_master_basico.repartition(100, "PAIS", "Familia")

# Guardar tabla básica para el notebook de cálculos
spark.sql("USE LH_INTELLIGENT_VISIBILITY_CARATULAS")

try:
    spark.sql("DROP TABLE IF EXISTS continental_master_basic_daily")
    print("Tabla anterior eliminada")
except:
    print("No había tabla anterior")

df_master_basico_optimizado.write.format("delta").mode("overwrite").saveAsTable("continental_master_basic_daily")

print("\n" + "="*80)
print("TABLA BÁSICA GUARDADA:")
print("="*80)
print("- Nombre: continental_master_basic_daily")
print("- Ubicación: LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_master_basic_daily")
print("- Contiene: Inv_Inicial, Inv_Final, Llegadas, Fcst, Facturación")
print("- CORRECCIÓN INVENTARIOS: Inv_Inicial = última semana mes anterior")
print("- CORRECCIÓN CSV: Cubos = 0 ahora usa CSV correctamente")
print("- Optimizada para cálculos avanzados")
print("="*80)

# ===============================================================================
# 10. VERIFICACIÓN FINAL
# ===============================================================================

print("\n=== VERIFICACIÓN FINAL ETL DIARIO ===")

print("\nSample CR-LAVA 2026 - DEBE TENER MAYO CON VALOR:")
df_master_basico.filter(
    (col("PAIS") == "CR") & (col("Familia") == "LAVA") & (col("Ano") == 2026)
).select("ORDENM", "Attribute", "Fcst", "Inv_Inicial", "Inv_Final") \
.orderBy("ORDENM").show(12, truncate=False)

print("\n=== VALIDACIÓN DE TRANSICIÓN DE AÑO ===")
print("Verificando continuidad Diciembre → Enero:")

# Mostrar transición de año si hay datos disponibles
df_validacion_transicion = df_master_basico.filter(
    (col("PAIS") == "CO") & (col("Familia") == "LAVA") &
    (((col("Ano") == ano_actual - 1) & (col("ORDENM") >= 11)) |  
     ((col("Ano") == ano_actual) & (col("ORDENM") <= 3)))        
).orderBy("Ano", "ORDENM").select("Ano", "ORDENM", "Attribute", "Inv_Inicial", "Inv_Final")

df_validacion_transicion.show(10, truncate=False)

# Verificación final de 2026
if csv_cargado:
    print("\n=== VERIFICACIÓN FINAL: FORECAST 2026 COMPLETO ===")
    
    # Contar meses con forecast en 2026
    meses_con_fcst_2026 = df_master_basico.filter(
        (col("Ano") == 2026) & (col("Fcst") > 0)
    ).select("ORDENM").distinct().count()
    
    # Calcular total forecast
    total_fcst_2026 = df_master_basico.filter(
        col("Ano") == 2026
    ).agg(spark_sum("Fcst").alias("Total")).first()["Total"]
    
    print(f"✅ 2026 tiene FORECAST en {meses_con_fcst_2026}/12 meses")
    print(f"✅ Total Forecast 2026: {total_fcst_2026:,.0f}")
    
    if meses_con_fcst_2026 == 12:
        print("✅ COBERTURA COMPLETA 2026 - CSV funcionando correctamente")

print("\n" + "="*80)
print("ETL DIARIO COMPLETADO EXITOSAMENTE")
print("="*80)
print("✅ INVENTARIOS: Corrección aplicada (LAG + transición de año)")
print("✅ CSV 2026: Lógica corregida (cubos=0 ahora usa CSV)")
print("✅ MAYO 2026: Ahora debería tener valor del CSV (2680.0)")
print("✅ PRIORIDAD: Cubos > 0 → CUBOS | Cubos = 0 → CSV")
print("\nLISTO PARA NOTEBOOK DE CÁLCULOS AVANZADOS")
print("="*80)
```

```python
#!/usr/bin/env python
# coding: utf-8

# ===============================================================================
# NOTEBOOK 1: CONTINENTAL_MASTER_ETL_DAILY
# Actualización diaria - Solo procesamiento básico y estable
# ===============================================================================

from datetime import datetime
from pyspark.sql.functions import col, year, month, max as spark_max, min as spark_min, count, lit, substring, coalesce, sum as spark_sum
from pyspark.sql.types import *
import pyspark.sql.functions as F

print("=== CONTINENTAL MASTER ETL DAILY ===")

# ===============================================================================
# 1. CONFIGURACIÓN GLOBAL (EXACTA DE TU CÓDIGO)
# ===============================================================================

# Usar SKUs ya definidos (saltando snippet)
# dict_skus ya debe estar disponible

# PARAMETROS GLOBALES (exacto de tu código)
fecha_actual = datetime.now()
ano_actual = fecha_actual.year
mes_actual_num = fecha_actual.month

# Mapeo de meses
nombres_meses = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril", 5: "Mayo", 6: "Junio",
    7: "Julio", 8: "Agosto", 9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

meses_mapping = {
    1: ("ENE", "Enero"), 2: ("FEB", "Febrero"), 3: ("MAR", "Marzo"), 
    4: ("ABR", "Abril"), 5: ("MAY", "Mayo"), 6: ("JUN", "Junio"),
    7: ("JUL", "Julio"), 8: ("AGO", "Agosto"), 9: ("SEP", "Septiembre"), 
    10: ("OCT", "Octubre"), 11: ("NOV", "Noviembre"), 12: ("DIC", "Diciembre")
}

mes_ejecucion_actual = nombres_meses[mes_actual_num]
anos_a_procesar = list(range(ano_actual, ano_actual + 6))

# Extraer SKUs válidos
skus_validos = []
for region, paises in dict_skus.items():
    for pais, familias in paises.items():
        for familia, skus in familias.items():
            skus_validos.extend(skus)

paises_requeridos = ["CL", "CO", "EC", "PE", "CR", "GT", "HN", "SV"]

print(f"Fecha actual del sistema: {fecha_actual.strftime('%Y-%m-%d')}")
print(f"Año actual: {ano_actual}")
print(f"Mes de ejecución: {mes_actual_num} ({mes_ejecucion_actual})")
print(f"Años a procesar: {anos_a_procesar}")
print(f"Total SKUs válidos: {len(skus_validos)}")
print(f"Países requeridos: {paises_requeridos}")

# ===============================================================================
# 2. CARGA DE DATOS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== CARGANDO TABLAS DEL LAKEHOUSE ===")

df_demanda = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_demanda")
df_entregas = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_entregas") 
df_costeado = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_costeado")
df_calendario = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_calendario")
df_compras = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_compras")
df_facturacion = spark.sql("SELECT * FROM LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_facturacion")

print("Tablas cargadas exitosamente")
print(f"  - continental_demanda: {df_demanda.count():,} registros")
print(f"  - continental_entregas: {df_entregas.count():,} registros")
print(f"  - continental_costeado: {df_costeado.count():,} registros")
print(f"  - continental_compras: {df_compras.count():,} registros")
print(f"  - continental_facturacion: {df_facturacion.count():,} registros")

# ===============================================================================
# 3. PREPARACIÓN DE DEMANDA (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO DATOS DE DEMANDA ===")

# Crear DataFrame de meses para cross join
meses_data = [(mes_num, codigo, nombre, mes_num) for mes_num, (codigo, nombre) in meses_mapping.items()]
df_meses = spark.createDataFrame(meses_data, ["ORDENM", "Attribute", "Mes_nombre", "mes_numero"])
df_anos = spark.createDataFrame([(ano,) for ano in anos_a_procesar], ["Ano"])

# Filtrar demanda solo con SKUs válidos desde año actual
df_demanda_prep = df_demanda.filter(
    (substring(col("Ejercicio_Periodo"), 1, 4).cast("int") >= ano_actual) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Pais").isNotNull()) &
    (col("Material").isin(skus_validos)) &
    (col("Grupo_Articulos") != "-")
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Plan_Demanda_a_Liberar").cast("double").alias("Fcst")
)

# Agrupar demanda por dimensiones principales
df_demanda_agg = df_demanda_prep.filter(
    col("Pais").isin(paises_requeridos)
).groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Fcst").alias("Fcst_total"),
    F.count("Material").alias("num_materiales")
)

print(f"Demanda agregada: {df_demanda_agg.count():,} registros")

# MATERIALIZAR DEMANDA
df_demanda_agg.cache()
df_demanda_agg.count()
print("Demanda materializada en cache")

# ===============================================================================
# 4. PREPARACIÓN DE FACTURACIÓN (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO DATOS DE FACTURACIÓN ===")

df_facturacion_prep = df_facturacion.filter(
    col("Material").isin(skus_validos)
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"), 
    col("Volumen_ventas").alias("Facturacion"),
    col("Ejercicio_Periodo"),
    substring(col("Ejercicio_Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio_Periodo"), 5, 3).cast("int").alias("mes_numero")
).filter(
    (col("Ano") >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
)

df_facturacion_agg = df_facturacion_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Facturacion").alias("Facturacion_total")
)

print(f"Facturación agregada: {df_facturacion_agg.count():,} registros")

# MATERIALIZAR FACTURACIÓN
df_facturacion_agg.cache()
df_facturacion_agg.count()
print("Facturación materializada en cache")

# ===============================================================================
# 5. PREPARACIÓN DE LLEGADAS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PREPARANDO LLEGADAS ===")

# Entregas
df_entregas_prep = df_entregas.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_entrega_confirmada")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_entrega_confirmada")).alias("Ano"),
    month(col("Fecha_entrega_confirmada")).alias("mes_numero"),
    col("Cantidad_Entregada_Efectivamente").alias("Entregas")
)

df_entregas_agg = df_entregas_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Entregas").alias("Entregas_total")
)

# Compras
df_compras_prep = df_compras.filter(
    (col("Material").isin(skus_validos)) &
    (year(col("Fecha_Final")) >= ano_actual) &
    (col("Pais").isin(paises_requeridos))
).select(
    col("Pais"),
    col("Grupo_articulos").alias("Familia"),
    col("Material"),
    year(col("Fecha_Final")).alias("Ano"),
    month(col("Fecha_Final")).alias("mes_numero"),
    col("Qty_sin_EE").alias("Compras")
)

df_compras_agg = df_compras_prep.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Compras").alias("Compras_total")
)

# Combinar entregas + compras = llegadas
df_entregas_union = df_entregas_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Entregas_total").alias("Cantidad")
)

df_compras_union = df_compras_agg.select(
    col("Pais"), col("Familia"), col("Ano"), col("mes_numero"),
    col("Compras_total").alias("Cantidad")
)

df_llegadas_combinado = df_entregas_union.union(df_compras_union)

df_llegadas_agg = df_llegadas_combinado.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    F.sum("Cantidad").alias("Llegadas_total")
)

print(f"Llegadas agregadas: {df_llegadas_agg.count():,} registros")

# MATERIALIZAR LLEGADAS
df_llegadas_agg.cache()
df_llegadas_agg.count()
print("Llegadas materializadas en cache")

# ===============================================================================
# 6. PROCESAMIENTO DE INVENTARIOS POR SEMANA (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== PROCESANDO INVENTARIOS POR SEMANA ===")

# AGREGAR por País+Familia+Semana
df_inventarios_por_semana = df_costeado.filter(
    (col("Sociedad_Pais").isin(paises_requeridos)) &
    (col("Grupo_Articulos").isNotNull()) &
    (col("Grupo_Articulos") != "-") &
    (substring(col("Ejercicio___Periodo"), 1, 4).cast("int") >= ano_actual)
).groupBy(
    col("Sociedad_Pais").alias("Pais"),
    col("Grupo_Articulos").alias("Familia"),
    substring(col("Ejercicio___Periodo"), 1, 4).cast("int").alias("Ano"),
    substring(col("Ejercicio___Periodo"), 5, 3).cast("int").alias("mes_numero"),
    col("Semana")
).agg(
    F.sum("Cantidad_Stock_Total_").alias("Inventario_semana")
)

# MATERIALIZAR INVENTARIOS
df_inventarios_por_semana.cache()
df_inventarios_por_semana.count()
print("Inventarios por semana materializados")

# Detectar primera y última semana por mes
df_limites_semanas = df_inventarios_por_semana.groupBy(
    "Pais", "Familia", "Ano", "mes_numero"
).agg(
    spark_min("Semana").alias("primera_semana"),
    spark_max("Semana").alias("ultima_semana")
)

# Inventario inicial = primera semana del mes
df_inv_inicial = df_inventarios_por_semana.alias("inv").join(
    df_limites_semanas.alias("lim"),
    (col("inv.Pais") == col("lim.Pais")) &
    (col("inv.Familia") == col("lim.Familia")) &
    (col("inv.Ano") == col("lim.Ano")) &
    (col("inv.mes_numero") == col("lim.mes_numero")) &
    (col("inv.Semana") == col("lim.primera_semana")),
    "inner"
).select(
    col("inv.Pais"), col("inv.Familia"), col("inv.Ano"), col("inv.mes_numero"),
    col("inv.Inventario_semana").alias("Inv_Inicial")
)

# Inventario final = última semana del mes
df_inv_final = df_inventarios_por_semana.alias("inv").join(
    df_limites_semanas.alias("lim"),
    (col("inv.Pais") == col("lim.Pais")) &
    (col("inv.Familia") == col("lim.Familia")) &
    (col("inv.Ano") == col("lim.Ano")) &
    (col("inv.mes_numero") == col("lim.mes_numero")) &
    (col("inv.Semana") == col("lim.ultima_semana")),
    "inner"
).select(
    col("inv.Pais"), col("inv.Familia"), col("inv.Ano"), col("inv.mes_numero"),
    col("inv.Inventario_semana").alias("Inv_Final")
)

# Combinar inicial y final
df_inventarios_mes = df_inv_inicial.join(
    df_inv_final,
    ["Pais", "Familia", "Ano", "mes_numero"],
    "inner"
)

print(f"Inventarios por mes: {df_inventarios_mes.count():,} registros")

# MATERIALIZAR INVENTARIOS MENSUALES
df_inventarios_mes.cache()
df_inventarios_mes.count()
print("Inventarios mensuales materializados")

# ===============================================================================
# 7. CREAR ESTRUCTURA BASE Y JOINS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== CREANDO ESTRUCTURA BASE ===")

# Obtener combinaciones únicas de País + Familia desde inventarios reales
paises_familias_inv = df_inventarios_mes.select("Pais", "Familia").distinct()

# Estructura base con inventarios
df_inventarios_master = paises_familias_inv \
    .crossJoin(df_anos) \
    .crossJoin(df_meses) \
    .alias("base") \
    .join(
        df_inventarios_mes.alias("inv"),
        (col("base.Pais") == col("inv.Pais")) &
        (col("base.Familia") == col("inv.Familia")) &
        (col("base.Ano") == col("inv.Ano")) &
        (col("base.ORDENM") == col("inv.mes_numero")),
        "left"
    ).select(
        col("base.Ano"),
        lit(mes_ejecucion_actual).alias("Mes"),
        col("base.Pais").alias("PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        lit(1).alias("PWOH"),
        F.coalesce(col("inv.Inv_Inicial"), F.lit(0.0)).alias("Inv_Inicial"),
        F.coalesce(col("inv.Inv_Final"), F.lit(0.0)).alias("Inv_Final")
    )

print(f"Estructura base con inventarios: {df_inventarios_master.count():,} registros")

# MATERIALIZAR ESTRUCTURA BASE
df_inventarios_master.cache()
df_inventarios_master.count()
print("Estructura base materializada")

# ===============================================================================
# 8. JOINS CON TODAS LAS MÉTRICAS (EXACTO DE TU CÓDIGO)
# ===============================================================================

print("=== UNIENDO TODAS LAS MÉTRICAS ===")

# JOIN con demanda
df_con_demanda = df_inventarios_master.alias("base") \
    .join(
        df_demanda_agg.alias("dem"),
        (col("base.PAIS") == col("dem.Pais")) &
        (col("base.Familia") == col("dem.Familia")) &
        (col("base.Ano") == col("dem.Ano")) &
        (col("base.ORDENM") == col("dem.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("dem.Fcst_total"), lit(0.0)).alias("Fcst")
    )

# JOIN con facturación
df_con_facturacion = df_con_demanda.alias("base") \
    .join(
        df_facturacion_agg.alias("fact"),
        (col("base.PAIS") == col("fact.Pais")) &
        (col("base.Familia") == col("fact.Familia")) &
        (col("base.Ano") == col("fact.Ano")) &
        (col("base.ORDENM") == col("fact.mes_numero")),
        "left"
    ).select(
        col("base.*"),
        coalesce(col("fact.Facturacion_total"), lit(0.0)).alias("Facturacion")
    )

# JOIN con llegadas
df_master_basico = df_con_facturacion.alias("base") \
    .join(
        df_llegadas_agg.alias("lleg"),
        (col("base.PAIS") == col("lleg.Pais")) &
        (col("base.Familia") == col("lleg.Familia")) &
        (col("base.Ano") == col("lleg.Ano")) &
        (col("base.ORDENM") == col("lleg.mes_numero")),
        "left"
    ).select(
        # Dimensiones base
        col("base.Ano"),
        col("base.Mes"),
        col("base.PAIS"),
        col("base.Familia"),
        col("base.Attribute"),
        col("base.ORDENM"),
        col("base.PWOH"),
        
        # Todas las métricas básicas
        col("base.Inv_Inicial"),
        col("base.Inv_Final"),
        coalesce(col("lleg.Llegadas_total"), lit(0.0)).alias("Llegadas"),
        col("base.Fcst"),
        col("base.Facturacion")
    )

print(f"Tabla maestra básica: {df_master_basico.count():,} registros")

# ===============================================================================
# 9. GUARDAR TABLA BÁSICA EN LAKEHOUSE
# ===============================================================================

print("=== GUARDANDO TABLA BÁSICA ===")

# Limpiar cache antes de guardar
spark.catalog.clearCache()

# Optimizar particionado
df_master_basico_optimizado = df_master_basico.repartition(100, "PAIS", "Familia")

# Guardar tabla básica para el notebook de cálculos
spark.sql("USE LH_INTELLIGENT_VISIBILITY_CARATULAS")

try:
    spark.sql("DROP TABLE IF EXISTS continental_master_basic_daily")
    print("Tabla anterior eliminada")
except:
    print("No había tabla anterior")

df_master_basico_optimizado.write.format("delta").mode("overwrite").saveAsTable("continental_master_basic_daily")

print("TABLA BÁSICA GUARDADA:")
print("- Nombre: continental_master_basic_daily")
print("- Ubicación: LH_INTELLIGENT_VISIBILITY_CARATULAS.continental_master_basic_daily")
print("- Contiene: Inv_Inicial, Inv_Final, Llegadas, Fcst, Facturación")
print("- Optimizada para cálculos avanzados")

# ===============================================================================
# 10. VERIFICACIÓN FINAL
# ===============================================================================

print("=== VERIFICACIÓN FINAL ETL DIARIO ===")

print("Sample CR-LAVA:")
df_master_basico.filter(
    (col("PAIS") == "CR") & (col("Familia") == "LAVA") & (col("Ano") == ano_actual)
).select("Attribute", "ORDENM", "Inv_Inicial", "Inv_Final", "Llegadas", "Fcst", "Facturacion") \
.orderBy("ORDENM").show(12, truncate=False)

print("ETL DIARIO COMPLETADO EXITOSAMENTE")
print("LISTO PARA NOTEBOOK DE CÁLCULOS AVANZADOS")
```
